```python
board=[' ',' ',' ',' ',' ',' ',' ',' ',' ']
symbol1=''
symbol2=''
Player1=''
Player2=''

def welcome():
    print("welcome to tic tac toe game")
    print("label are as shown\n")
    print('Symbols are O and X')
    print('7|8|9\n--------\n4|5|6\n--------\n1|2|3')
def playerinfo():
    global symbol1
    global symbol2
    global Player1
    global Player2
    Player1 = str(input('enter the name of the player 1: '))
    
    symbol1 = str(input('enter the symbol of player1: ')).upper()
    Player2 = str(input('enter the name of the player 2: '))
    if symbol1=='X':
        symbol2 ='O'
    else:
        symbol2='X'

def enterpos(sym1,sym2):
    global board
    win=""
    while(True):
        pos = int(input('Player 1 turn: '))
        pattern(pos,sym1)
        if checkpattern(board,sym1) == True:
            win='player 1 win'
            break
        pos = int(input('player 2 turn: '))
        pattern(pos,sym2)
        if checkpattern(board,sym2) == True:
            win='player 2 win'
            break
        
    
def pattern(num,sym):
    global board
    board[num-1] = sym
    print(f'{board[6]}|{board[7]}|{board[8]}\n---------------\n{board[3]}|{board[4]}|{board[5]}\n---------------\n{board[0]}|{board[1]}|{board[2]}\n')  
def checkpattern(board,mark):
    flag=False
    for i in range(0,9,2):
        if board[i]==board[i+1]==board[i+2]==mark:
            flag=True
            break
    for i in range(0,3):
        if board[i]==board[i+3]==board[i+6]==mark:
            flag=True
            break
    if board[0]==board[4]==board[8]==mark:
        flag=True
        break
    if board[2]==board[4]==board[6]==mark:
        flag=True
        break
return flag
    
welcome()
playerinfo()    
enterpos(symbol1,symbol2)

    
    
    
    
    
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
N = int(input())
lst = []
num=0
maxcountlist=[]
#mean program
num = input()
for item in num.split():
    lst.append(int(item))
print("{}".format(sum(lst)/N))
#meadian program
lst.sort()

sorted_list = lst
if N % 2 == 0 :
    pos = N//2
    res=(sorted_list[pos-1]+sorted_list[pos])/2
    print("{}".format(res))
    
else:
    pos = N//2 +1
    print("{}".format(sorted_list[pos-1]))
# #mode program
dic = {}


for item in set(sorted_list):
    dic[item]=sorted_list.count(item)
maxcount =max(dic.values())
for item in dic.keys():
     if dic.get(item)==maxcount:
            maxcountlist.append(item)
print(min(maxcountlist))
    
```

```python
import random

board = [' ']*9
symbol1 = ''
symbol2 = ''
counter=0
firstplayer=None
play=False
status=False 


def playfirst():
    temp =random.randint(1,10)
    if temp%2==0:
        return True
    else:
        return False
    
def displayBoard(board):
    print(f'{board[6]}|{board[7]}|{board[8]}\n{board[3]}|{board[4]}|{board[5]}\n{board[0]}|{board[1]}|{board[2]}')
    
def isfull(board):
    stat = None
    for item in board:
        if item == ' ':
            stat = False
            break            
        else:
            stat = True
    return stat
        
def posempty(pos,board):
    
    if board[pos] == ' ':
        return True
    else:
        return False
    
def putmark(val,symbol):
    global board
    board[val]=symbol
    
def checkstatus(board):
    win=''
    stat=False
    for i in range(0,9,3):
        if board[i]==board[i+1]==board[i+2]=='X':
            win = 'player 1 win'
            stat= True
            break
        elif board[i]==board[i+1]==board[i+2]=='O':
            win = 'player 2 win'
            stat= True
            break
        else:
            pass
            
    for i in range(0,3):
        if board[i]==board[i+3]==board[i+6]=='X':
            win = 'player 1 win'
            stat= True
            break
        elif board[i]==board[i+3]==board[i+6]=='O':
            win = 'player 2 win'
            stat= True
            break
        else:
            pass
            
    if board[0]==board[4]==board[8]=='X'or board[2]==board[4]==board[6]=='X':
        win = 'player 1 win'
        stat= True
        
    elif board[0]==board[4]==board[8]=='O'or board[2]==board[4]==board[6]=='O':
        win = 'player 2 win'
        stat= True
        
    else:
        pass
        
    if(isfull(board)==True):
        if stat == False:
            win = 'Match draw'
            stat= True
        else:
            pass
    else:
        pass
    
    if stat == True:
        print(win)
        return stat
    else:
        return stat
    
def getinput(symbol1,symbol2):
    global board
    global counter
    while True:
        while True:
            try:
                val = int(input('choose pos between 1 to 9: '))

            except:
                print('please enter a valid value')
            else:
                break

        if posempty(val-1,board)==True or val > 9:
            if(counter%2==0):
                    putmark(val-1,symbol2)
            else:
                    putmark(val-1,symbol1)
            counter+=1
            break
        else:
            print('Position already marked or pos invalid.\nEnter valid value again')
            continue
    displayBoard(board)
    return checkstatus(board)
        
        
        
while play != 'n':
    print('welcome to tic tac toe')
    while True:
        symbol1 = input('player 1: Do you want to be X or O:').upper()
        if symbol1 == 'X':
            symbol2='O'
            break
        elif symbol1 == 'O':
            symbol2='X'
            break
        else:
            print('Wrong input')
    firstplayer=playfirst()
    print(f'symbol1 = {symbol1} symbol2 = {symbol2} firstplayer = {firstplayer}')
    if firstplayer == True:
        print('player 1 go first')
        counter = 1
    else:
        print('player 2 play first')
        counter = 2
    while status!=True:
        status = getinput(symbol1,symbol2)
        print(status)
    play = input('Do you want to play again(y/n):')
    while True:
        if play=='y':
            board = [' '] * 9
            status= False
            break
        else:
            pass
        
    
```
